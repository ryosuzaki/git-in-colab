# Initial setting

```python
# Initial setting
# https://stackoverflow.com/questions/59454990/how-to-push-from-colab-to-github
from google.colab import drive
import os

# Mount Google Drive
drive.mount('/content/drive')

# Set initial information
mount_dir="/content/drive/MyDrive"
# Directory to put repositories in
repos_name="git_repos"
repos_dir=os.path.join(mount_dir,repos_name)

# GitHub email username
email="your_email@example.com"
uname="your_username"

# Personal access token https://docs.github.com/en/authentication/keeping-your-account-and-data-secure/managing-your-personal-access-tokens
# https://github.com -> settings -> Developer settings -> Personal access tokens -> Generate
# Check repo in scopes
token="your_personal_access_token"

# If the directory does not exist, create it
%mkdir -p $repos_dir
# Set config
!git config --global user.email $email
!git config --global user.name $uname
```

# Template

```python
# Set the repository to handle
repo_name="target_repo"
repo_dir=os.path.join(repos_dir,repo_name)
url=f"https://{uname}:{token}@github.com/{uname}/{repo_name}"
```

```python
# Clone
%cd $repos_dir
!git clone $url
```

```python
# Pull
%cd $repo_dir
!git pull
```

```python
# Commit
%cd $repo_dir
# https://qiita.com/konweb/items/621722f67fdd8f86a017
!git add .
!git commit -a -m "comment"
```

```python
# Push
%cd $repo_dir
!git push $url
```
