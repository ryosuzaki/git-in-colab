```python
# https://stackoverflow.com/questions/59454990/how-to-push-from-colab-to-github
from google.colab import drive
import os

# Mount Google Drive
drive.mount('/content/drive')

# Set initial information
drive="/content/drive/MyDrive"
# Directory to put repositories in
repos_name="git_repos"
repos_dir=os.path.join(drive,repos_name)
# Target repository
repo_name="target_repo"
repo_dir=os.path.join(repos_dir,repo_name)

# GitHub email user
email="your_email@example.com"
uname="your_username"
# Personal access token
# https://github.com -> settings -> Developer settings -> Personal access tokens -> Generate
# Check repo in scopes
token="your_personal_access_token"

# Initial setting
%mkdir -p $repos_dir
!git config --global user.email $email
!git config --global user.name $uname
```

```python
#clone
%cd $repos_dir
url=f"https://{uname}:{token}@github.com/{uname}/{repo_name}"
!git clone $url
```

```python
# Commit
%cd $repo_dir
#https://qiita.com/konweb/items/621722f67fdd8f86a017
!git add .
!git commit -a -m "add main"
```

```python
# Push
%cd $repo_dir
url=f"https://{token}@github.com/{uname}/{repo_name}"
!git push $url
```

```python
# Download git-lfs to Push Files larger than 100MB.
!wget -O git-lfs.tar.gz https://github.com/git-lfs/git-lfs/releases/download/v2.13.2/git-lfs-linux-amd64-v2.13.2.tar.gz
!tar xzf git-lfs.tar.gz
!bash ./install.sh
!git lfs install
%cd $repo_dir
#FILE_NAME is the file with size >100MB and you wants to PUSH to GITHUB
!git lfs track <FILE_NAME>
```
